```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Создаем синтетические данные с тремя кластерами
data, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Используем метод "правило локтя" для определения оптимального количества кластеров
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Построим график
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Метод "правило локтя" для определения количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(range(1, 11))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

# Создаем синтетические данные с тремя кластерами
data, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Используем коэффициент силуэта для определения оптимального количества кластеров
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    cluster_labels = kmeans.fit_predict(data)
    silhouette_avg = silhouette_score(data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Построим график
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Коэффициент силуэта для определения количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage

# Создаем искусственный набор данных для примера
X, y = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Иерархическая кластеризация
linked = linkage(X, 'ward')

# Строим дендрограмму
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
