```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats


data = pd.read_csv("insurance.csv")


bmi_data = data['bmi']
charges_data = data['charges']

bmi_mode = scipy.stats.mode(bmi_data, keepdims=True).mode[0]
bmi_median = bmi_data.median()
bmi_mean = bmi_data.mean()
bmi_range = bmi_data.max() - bmi_data.min()
bmi_std = bmi_data.std()
bmi_iqr = scipy.stats.iqr(bmi_data)

charges_mode = scipy.stats.mode(charges_data, keepdims=True).mode[0]
charges_median = charges_data.median()
charges_mean = charges_data.mean()
charges_range = charges_data.max() - charges_data.min()
charges_std = charges_data.std()
charges_iqr = scipy.stats.iqr(charges_data)

print("      BMI:")
print(f": {bmi_mode}")
print(f": {bmi_median}")
print(f": {bmi_mean}")
print(f": {bmi_range}")
print(f" : {bmi_std}")
print(f" : {bmi_iqr}")

print("\n      Charges:")
print(f": {charges_mode}")
print(f": {charges_median}")
print(f": {charges_mean}")
print(f": {charges_range}")
print(f" : {charges_std}")
print(f" : {charges_iqr}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(bmi_data, bins=30, color='skyblue', edgecolor='black')
plt.axvline(bmi_mode, color='r', linestyle='dashed', linewidth=2, label=f': {bmi_mode}')
plt.axvline(bmi_median, color='g', linestyle='dashed', linewidth=2, label=f': {bmi_median}')
plt.axvline(bmi_mean, color='b', linestyle='dashed', linewidth=2, label=f': {bmi_mean}')
plt.title('  BMI')
plt.xlabel('BMI')
plt.ylabel('')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(charges_data, bins=30, color='skyblue', edgecolor='black')
plt.axvline(charges_mode, color='r', linestyle='dashed', linewidth=2, label=f': {charges_mode}')
plt.axvline(charges_median, color='g', linestyle='dashed', linewidth=2, label=f': {charges_median}')
plt.axvline(charges_mean, color='b', linestyle='dashed', linewidth=2, label=f': {charges_mean}')
plt.title('  Charges')
plt.xlabel('')
plt.ylabel('')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("insurance.csv")

selected_feature = "bmi"

num_samples = 300

sample_sizes = [1000, 3000, 5000, 10000]

means = []

#      
for n in sample_sizes:
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data[selected_feature], size=n, replace=True)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)
    means.append(sample_means)

#     
plt.figure(figsize=(12, 8))
for i, n in enumerate(sample_sizes):
    plt.subplot(2, 2, i + 1)
    plt.hist(means[i], bins=30, density=True)
    plt.title(f"Distribution of Sample Means (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

#         
for i, n in enumerate(sample_sizes):
    sample_std = np.std(means[i])
    sample_mean = np.mean(means[i])
    print(f"  n={n}:")
    print(f": {sample_mean:.2f}")
    print(f" : {sample_std:.2f}")
    print()

#  
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv('insurance.csv')

charges = data['charges']
bmi = data['bmi']


confidence_level_95 = 0.95
confidence_level_99 = 0.99


mean_charges = np.mean(charges)
std_dev_charges = np.std(charges, ddof=1)  
n_charges = len(charges)

#       "bmi"
mean_bmi = np.mean(bmi)
std_dev_bmi = np.std(bmi, ddof=1) 
n_bmi = len(bmi)

#   
#  "charges"
margin_of_error_95_charges = stats.t.ppf((1 + confidence_level_95) / 2, df=n_charges - 1) * (std_dev_charges / np.sqrt(n_charges))
margin_of_error_99_charges = stats.t.ppf((1 + confidence_level_99) / 2, df=n_charges - 1) * (std_dev_charges / np.sqrt(n_charges))

confidence_interval_95_charges = (mean_charges - margin_of_error_95_charges, mean_charges + margin_of_error_95_charges)
confidence_interval_99_charges = (mean_charges - margin_of_error_99_charges, mean_charges + margin_of_error_99_charges)

#  "bmi"
margin_of_error_95_bmi = stats.t.ppf((1 + confidence_level_95) / 2, df=n_bmi - 1) * (std_dev_bmi / np.sqrt(n_bmi))
margin_of_error_99_bmi = stats.t.ppf((1 + confidence_level_99) / 2, df=n_bmi - 1) * (std_dev_bmi / np.sqrt(n_bmi))

confidence_interval_95_bmi = (mean_bmi - margin_of_error_95_bmi, mean_bmi + margin_of_error_95_bmi)
confidence_interval_99_bmi = (mean_bmi - margin_of_error_99_bmi, mean_bmi + margin_of_error_99_bmi)

#  
print("95%      (charges):", confidence_interval_95_charges)
print("99%      (charges):", confidence_interval_99_charges)

print("95%        (bmi):", confidence_interval_95_bmi)
print("99%        (bmi):", confidence_interval_99_bmi)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

#     CSV
data = pd.read_csv('insurance.csv')

#   "bmi"  "charges"
bmi = data['bmi']
charges = data['charges']

# - 
ks_stat_bmi, p_value_bmi = stats.kstest(bmi, 'norm', args=(bmi.mean(), bmi.std()))
ks_stat_charges, p_value_charges = stats.kstest(charges, 'norm', args=(charges.mean(), charges.std()))

print(f"KS-     (bmi):")
print(f"KS-: {ks_stat_bmi}")
print(f"P-: {p_value_bmi}")
print()

print(f"KS-   (charges):")
print(f"KS-: {ks_stat_charges}")
print(f"P-: {p_value_charges}")
print()

# Q-Q 
def plot_qq(data, title):
    plt.figure(figsize=(8, 4))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.show()

plot_qq(bmi, "Q-Q      (bmi)")
plot_qq(charges, "Q-Q    (charges)")

# 
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(bmi, kde=True)
plt.title("    (bmi)")

plt.subplot(1, 2, 2)
sns.histplot(charges, kde=True)
plt.title("  (charges)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

def compare_bmi_means(data_file):
  
    data = pd.read_csv(data_file, sep=',')
    
    data_northwest = data[data['region'] == 'northwest']['bmi']
    data_southwest = data[data['region'] == 'southwest']['bmi']

    #   
    shapiro_northwest = stats.shapiro(data_northwest)
    shapiro_southwest = stats.shapiro(data_southwest)
    
    #    
    bartlett_result = stats.bartlett(data_northwest, data_southwest)
    
    #      
    t_stat, p_value = stats.ttest_ind(data_northwest, data_southwest)

    print("  :")
    print("Northwest p-value =", shapiro_northwest.pvalue)
    print("Southwest p-value =", shapiro_southwest.pvalue)
    if shapiro_northwest.pvalue > 0.05 and shapiro_southwest.pvalue > 0.05:
        print("   .")
    else:
        print("       .")
    
    print("\n   :")
    print("Bartlett p-value =", bartlett_result.pvalue)
    if bartlett_result.pvalue > 0.05:
        print("   .")
    else:
        print("  .")
    
    print("\n     :")
    print("t-statistic =", t_stat)
    print("p-value =", p_value)
    if p_value < 0.05:
        print("  ,   .")
    else:
        print("   ,    .")

    #   
    if p_value < 0.05:
        print(":   BMI   .")
    else:
        print(":   BMI  .")

#  :
compare_bmi_means('bmi.csv')
```

```python
import pandas as pd
import scipy.stats as sts

def main():
    data = pd.DataFrame({'Side': [1, 2, 3, 4, 5, 6], 'Counts': [97, 98, 109, 95, 97, 104]})
    print(data)
    
    observed_counts = data['Counts']
    expected_counts = [100] * 6  #       ( )
    
    print('         -')
    chi_square_statistic, p_value = sts.chisquare(f_obs=observed_counts, f_exp=expected_counts)
    print('Chi-Square Statistic:', chi_square_statistic)
    print('p-value:', p_value)

    if p_value > 0.05:
        print(':   .')
    else:
        print(':    .')

if __name__ == '__main__':
    main()
```

```python
import pandas as pd
import scipy.stats as stats

#  DataFrame
data = pd.DataFrame({
    '': [89, 17, 11, 43, 22, 1],
    ' ': [80, 22, 20, 35, 6, 4],
    '   ': [35, 44, 35, 6, 8, 22]
})
data.index = ['  ', ' ', '  ', ' ', ' ', '']

#   -
chi2, p, dof, expected = stats.chi2_contingency(data)

#  
print('Chi-Square Statistic:', chi2)
print('p-value:', p)
print(' :', dof)
print(' :', expected)

if p < 0.05:
    print(':      (   ).')
else:
    print(':       (   ).')
```
