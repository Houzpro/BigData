```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("insurance.csv")

selected_feature = "bmi"

num_samples = 300

sample_sizes = [1000, 3000, 5000, 10000]

means = []

# Генерация выборок и вычисление средних значений
for n in sample_sizes:
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data[selected_feature], size=n, replace=True)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)
    means.append(sample_means)

# Построение гистограмм для средних значений
plt.figure(figsize=(12, 8))
for i, n in enumerate(sample_sizes):
    plt.subplot(2, 2, i + 1)
    plt.hist(means[i], bins=30, density=True)
    plt.title(f"Distribution of Sample Means (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

# Рассчет стандартного отклонения и среднего для каждой длины выборки
for i, n in enumerate(sample_sizes):
    sample_std = np.std(means[i])
    sample_mean = np.mean(means[i])
    print(f"Длина выборки n={n}:")
    print(f"Среднее: {sample_mean:.2f}")
    print(f"Стандартное отклонение: {sample_std:.2f}")
    print()

# Вывод результатов
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv('insurance.csv')

charges = data['charges']
bmi = data['bmi']


confidence_level_95 = 0.95
confidence_level_99 = 0.99


mean_charges = np.mean(charges)
std_dev_charges = np.std(charges, ddof=1)  
n_charges = len(charges)

# Вычисление среднего и стандартного отклонения для "bmi"
mean_bmi = np.mean(bmi)
std_dev_bmi = np.std(bmi, ddof=1) 
n_bmi = len(bmi)

# Вычисление доверительных интервалов
# Для "charges"
margin_of_error_95_charges = stats.t.ppf((1 + confidence_level_95) / 2, df=n_charges - 1) * (std_dev_charges / np.sqrt(n_charges))
margin_of_error_99_charges = stats.t.ppf((1 + confidence_level_99) / 2, df=n_charges - 1) * (std_dev_charges / np.sqrt(n_charges))

confidence_interval_95_charges = (mean_charges - margin_of_error_95_charges, mean_charges + margin_of_error_95_charges)
confidence_interval_99_charges = (mean_charges - margin_of_error_99_charges, mean_charges + margin_of_error_99_charges)

# Для "bmi"
margin_of_error_95_bmi = stats.t.ppf((1 + confidence_level_95) / 2, df=n_bmi - 1) * (std_dev_bmi / np.sqrt(n_bmi))
margin_of_error_99_bmi = stats.t.ppf((1 + confidence_level_99) / 2, df=n_bmi - 1) * (std_dev_bmi / np.sqrt(n_bmi))

confidence_interval_95_bmi = (mean_bmi - margin_of_error_95_bmi, mean_bmi + margin_of_error_95_bmi)
confidence_interval_99_bmi = (mean_bmi - margin_of_error_99_bmi, mean_bmi + margin_of_error_99_bmi)

# Вывод результатов
print("95% Доверительный интервал для среднего расходов (charges):", confidence_interval_95_charges)
print("99% Доверительный интервал для среднего расходов (charges):", confidence_interval_99_charges)

print("95% Доверительный интервал для среднего индекса массы тела (bmi):", confidence_interval_95_bmi)
print("99% Доверительный интервал для среднего индекса массы тела (bmi):", confidence_interval_99_bmi)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузите данные из файла CSV
data = pd.read_csv('insurance.csv')

# Выберите признаки "bmi" и "charges"
bmi = data['bmi']
charges = data['charges']

# Колмогорова-Смирнова тест
ks_stat_bmi, p_value_bmi = stats.kstest(bmi, 'norm', args=(bmi.mean(), bmi.std()))
ks_stat_charges, p_value_charges = stats.kstest(charges, 'norm', args=(charges.mean(), charges.std()))

print(f"KS-тест для индекса массы тела (bmi):")
print(f"KS-статистика: {ks_stat_bmi}")
print(f"P-значение: {p_value_bmi}")
print()

print(f"KS-тест для расходов (charges):")
print(f"KS-статистика: {ks_stat_charges}")
print(f"P-значение: {p_value_charges}")
print()

# Q-Q графики
def plot_qq(data, title):
    plt.figure(figsize=(8, 4))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(title)
    plt.show()

plot_qq(bmi, "Q-Q график для индекса массы тела (bmi)")
plot_qq(charges, "Q-Q график для расходов (charges)")

# Визуализация
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(bmi, kde=True)
plt.title("Распределение индекса массы тела (bmi)")

plt.subplot(1, 2, 2)
sns.histplot(charges, kde=True)
plt.title("Распределение расходов (charges)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

def compare_bmi_means(data_file):
  
    data = pd.read_csv(data_file, sep=',')
    
    data_northwest = data[data['region'] == 'northwest']['bmi']
    data_southwest = data[data['region'] == 'southwest']['bmi']

    # Проверка на нормальность
    shapiro_northwest = stats.shapiro(data_northwest)
    shapiro_southwest = stats.shapiro(data_southwest)
    
    # Проверка на гомогенность дисперсии
    bartlett_result = stats.bartlett(data_northwest, data_southwest)
    
    # Сравнение средних значений по критерию Стьюдента
    t_stat, p_value = stats.ttest_ind(data_northwest, data_southwest)

    print("Проверка на нормальность:")
    print("Northwest p-value =", shapiro_northwest.pvalue)
    print("Southwest p-value =", shapiro_southwest.pvalue)
    if shapiro_northwest.pvalue > 0.05 and shapiro_southwest.pvalue > 0.05:
        print("Выборки имеют нормальное распределение.")
    else:
        print("Одна или обе выборки не имеют нормальное распределение.")
    
    print("\nПроверка на гомогенность дисперсии:")
    print("Bartlett p-value =", bartlett_result.pvalue)
    if bartlett_result.pvalue > 0.05:
        print("Дисперсии выборок примерно одинаковы.")
    else:
        print("Дисперсии выборок различны.")
    
    print("\nСравнение средних значений по критерию Стьюдента:")
    print("t-statistic =", t_stat)
    print("p-value =", p_value)
    if p_value < 0.05:
        print("Нулевая гипотеза отвергается, средние значения различны.")
    else:
        print("Нулевая гипотеза не отвергается, средние значения примерно равны.")

    # Добавим текстовое описание
    if p_value < 0.05:
        print("Вывод: Средние значения BMI различаются статистически значимо.")
    else:
        print("Вывод: Средние значения BMI примерно равны.")

# Пример использования:
compare_bmi_means('bmi.csv')
```

```python
import pandas as pd
import scipy.stats as sts

def main():
    data = pd.DataFrame({'Side': [1, 2, 3, 4, 5, 6], 'Counts': [97, 98, 109, 95, 97, 104]})
    print(data)
    
    observed_counts = data['Counts']
    expected_counts = [100] * 6  # Ожидаемые равные вероятности для каждой стороны (равномерное распределение)
    
    print('Проверка распределения на равномерность с помощью встроенной функции критерия Хи-квадрата')
    chi_square_statistic, p_value = sts.chisquare(f_obs=observed_counts, f_exp=expected_counts)
    print('Chi-Square Statistic:', chi_square_statistic)
    print('p-value:', p_value)

    if p_value > 0.05:
        print('Вывод: Распределение является равномерным.')
    else:
        print('Вывод: Распределение не является равномерным.')

if __name__ == '__main__':
    main()
```

```python
import pandas as pd
import scipy.stats as stats

# Создать DataFrame
data = pd.DataFrame({
    'Женат': [89, 17, 11, 43, 22, 1],
    'Гражданский брак': [80, 22, 20, 35, 6, 4],
    'Не состоит в отношениях': [35, 44, 35, 6, 8, 22]
})
data.index = ['Полный рабочий день', 'Частичная занятость', 'Временно не работает', 'На домохозяйстве', 'На пенсии', 'Учёба']

# Выполнить критерий хи-квадрат
chi2, p, dof, expected = stats.chi2_contingency(data)

# Вывести результаты
print('Chi-Square Statistic:', chi2)
print('p-value:', p)
print('Степени свободы:', dof)
print('Ожидаемые частоты:', expected)

if p < 0.05:
    print('Вывод: Семейное положение влияет на занятость (существует статистически значимая связь).')
else:
    print('Вывод: Семейное положение не влияет на занятость (нет статистически значимой связи).')
```
